# Feature Engineering 

### i.) Missing Value Treatment

```python
# Missing values in Categorical Data. As we saw in the above graphs with frequency on y axis, most of the features which have 
# NaN can be replaced using MODE value.
```

```python
total = df[categorical_feat].isnull().sum().sort_values(ascending = False)
percent = (df[categorical_feat].isnull().mean()*100).sort_values(ascending=False)
miss_category = pd.concat([total,percent],axis=1,keys=['Total','Percent'])
miss_category.head(17)
```

```python
# Let's drop the features with more than 45% of missing data.
def drop_high_miss(df):
    print('Before dropping : ',df.shape)
    df.drop((miss_category[miss_category['Percent'] > 6]).index,1,inplace=True)
    print('After dropping : ',df.shape)
```

```python
drop_high_miss(df)
```

```python
# Update the miss_category and categorical_feat for future use.
```

```python
miss_category = miss_category[miss_category['Percent']<10]
```

```python
category_feat_after_drop = list(miss_category.index.values)
print(len(category_feat_after_drop))
print(category_feat_after_drop)
```

```python
def fill_na_mode(df):
    for feature in category_feat_after_drop:
        if df[feature].isnull().sum() > 0:
            df[feature]=df[feature].fillna(df[feature].mode()[0])
        else:
            pass
```

```python
fill_na_mode(df)
```

```python
# missing values in continuous_numeric
```

```python
total = df[continuous_numeric].isnull().sum().sort_values(ascending = False)
percent = (df[continuous_numeric].isnull().mean()*100).sort_values(ascending=False)
miss_cont_numeric = pd.concat([total,percent],axis=1,keys=['Total','Percent'])
miss_cont_numeric.head(3)
```

```python
def cont_num_fillna(df):
    for feature in continuous_numeric:
        if df[feature].isnull().sum() > 0:
            df[feature] = df[feature].fillna(df[feature].mean())
        else:
            pass
```

```python
cont_num_fillna(df)
```

```python
# Missning values in discrete numeric
```

```python
total = df[discrete_numeric].isnull().sum().sort_values(ascending = False)
percent = (df[discrete_numeric].isnull().mean()*100).sort_values(ascending=False)
miss_disc_numeric = pd.concat([total,percent],axis=1,keys=['Total','Percent'])
miss_disc_numeric.head(2)
```

```python
# Missning values in Year features
```

```python
total = df[year_feat].isnull().sum().sort_values(ascending = False)
percent = (df[year_feat].isnull().mean()*100).sort_values(ascending=False)
miss_year_feat = pd.concat([total,percent],axis=1,keys=['Total','Percent'])
miss_year_feat.head(2)
```

```python
def year_feat_fillna_median(df):
    for feature in year_feat:
        if df[feature].isnull().sum() > 0:
            df[feature] = df[feature].fillna(df[feature].median())
        else:
            pass
```

```python
year_feat_fillna_median(df)
```

```python
df.isnull().sum().max()
```

```python
# Thus, now no missing data.
```

### ii.) Feature Extraction

```python
# Let's transform the features we saw above:
# SalePrice
# LotFrontage
# LotArea
# 1stFlrSF
# GrLivArea
```

```python
data = df.copy()
def log_transf(data,feature):
    data['log_'+feature] = np.log(data[feature])
    data.drop([feature],axis=1,inplace=True)
```

```python
log_transf(data,'SalePrice')
```

```python
log_transf(data,'LotFrontage')
log_transf(data,'LotArea')
log_transf(data,'1stFlrSF')
log_transf(data,'GrLivArea')
```

```python
data.head()
```

```python
# Now we can add the Year calculated features:
# YrSold-YearBuilt
# YrSold-GarageYrBlt   
# YrSold-YearRemodAdd
```

```python
for feature in ['YearBuilt','GarageYrBlt','YearRemodAdd']:
    data['calc_'+feature] = data['YrSold'] - data[feature]
    data.drop([feature],axis=1,inplace=True)
```

```python
data.head()
```

```python
# Now the only thing left is to convert categorical values to numeric.
# Here, pd.get_dummies would lead to curse of dimensionality. Thus, using Labeled Encoder
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
for feature in category_feat_after_drop:
    le = LabelEncoder()
    data[feature] = le.fit_transform(data[feature])
```
